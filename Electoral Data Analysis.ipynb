Q5
Determine the consistently tri-polar constituencies. I.e Constituencies in which 3
parties or more got greater than 10% of the vote shares in each election. The
parties can be different for different elections.

```python
q5_data = data[['AC No','Election Year','Votes %']]
```

Converted Votes % column from string to float so it can be benchmarked against the 10% of the vote shares in the election.

Considered votes % received by candidate and compared it to 10%, and candidates with more than 10% were considered for futher analysis

```python
q5_data = q5_data.replace(to_replace = r"\%", value = "", regex = True)
q5_data['Votes %'] = q5_data['Votes %'].astype(float)
```

```python
q5_data
```

```python
q5_data = q5_data.groupby(by=['AC No','Election Year']).head(3)
```

```python
q5_data['Top1'] = q5_data['Votes %'][::3]
q5_data['Top2'] = q5_data['Votes %'][1::3]
q5_data['Top3'] = q5_data['Votes %'][2::3]
```

```python
q5_data.head(15)
```

```python
q5_data_fillna = q5_data.fillna(method = 'ffill')
q5_data_fillna.head(10)
```

```python
q5_data_final = q5_data_fillna[2::3]
q5_data_final
```

```python
q5_data_final = q5_data_final.drop(columns = ['Votes %'])
q5_data_final.set_index('AC No', inplace = True)
q5_data_final
```

```python
q5_data_final['Top3 more than 10'] = np.where(((q5_data_final['Top1'] > 10 ) & (q5_data_final['Top2'] > 10) & (q5_data_final['Top3'] > 10)),1,0)
```

Every constituency in every election year with their top 3 candidates percentage.

Whichever combination of election year and constituency had their top 3 percentages > 10% were considered for further analysis

```python
q5_data_final.groupby('Election Year').head()
```

Whichever constituency had their top 3 percentages over 10, across 4 years of election were considered for further analysis.

```python
q5_pivot = q5_data_final.pivot_table(index = ['AC No'], columns = 'Election Year', values = 'Top3 more than 10')
q5_pivot
```

```python
q5_pivot['Tri-Polar'] = np.where((q5_pivot['2005-Feb']==1) & (q5_pivot['2005-Oct']==1) & (q5_pivot['2010']==1) & (q5_pivot['2015']==1),'Yes','No')
```

```python
q5_pivot.sort_values(by = 'Tri-Polar', ascending = False, inplace = True)
q5_answer = q5_pivot.head(27)
q5_answer.reset_index(inplace=True)
q5_answer
```

```python
q5_answer_final = pd.merge(q5_answer, name_and_no, on ='AC No', how = 'left')
q5_answer_final
```
